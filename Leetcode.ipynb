## Reverse Vowels of a String

```python
def reverseVowels(s):
    """
    :type s: str
    :rtype: str
    """
    vowels = "aeiouAEIOU"
    s, i, j = list(s), 0, len(s) - 1
    while i < j:
        if not s[i] in vowels:
            i += 1
        elif not s[j] in vowels:
            j -= 1
        else:
            s[i], s[j] = s[j], s[i]
            i, j = i + 1, j - 1
    return "".join(s)
```

```python
reverseVowels("hello")
```

## Reverse String 

```python
def reverseString(s):
    """
    :type s: List[str]
    :rtype: None Do not return anything, modify s in-place instead.
    """
    i, j = 0, len(s) - 1
    while i < j:
        s[i], s[j] = s[j], s[i]
        i, j = i + 1, j - 1
```

```python
mystr = ["h","e","l","l","o"]
reverseString(mystr)
print(mystr)
```

## Reverse only letters

```python
def reverseOnlyLetters(s):
    """
    :type s: str
    :rtype: str
    """
    s, i, j = list(s), 0, len(s) - 1
    while i < j:
        if not s[i].isalpha():
            i += 1
        elif not s[j].isalpha():
            j -= 1
        else:
            s[i], s[j] = s[j], s[i]
            i, j = i + 1, j - 1
    return "".join(s)
```

```python
reverseOnlyLetters("ab-cd-12-llak")
```

## A Number After Double Reversal

```python
def isSameAfterReversals(num):
    """
    :type num: int
    :rtype: bool
    """
    return num == 0 or num % 10
```

```python
isSameAfterReversals(18000)
```

```python
isSameAfterReversals(2341)
```

## Length of Last Word

```python
def lengthOfLastWord(s):
    """
    :type s: str
    :rtype: int
    """
    S = s.strip()
    arrr = S.split()
    return len(arrr[-1])
```

```python
lengthOfLastWord("Hello World")
```

```python
lengthOfLastWord("    length of last word    ")
```

## Power of 2

```python
def isPowerOfTwo(n):
    """
    :type n: int
    :rtype: bool
    """
    if n <= 0:
        return False
    while n % 2 == 0:
        n /= 2
    return n == 1
```

```python
isPowerOfTwo(16)
```

```python
isPowerOfTwo(11)
```
