## 1 to N without using loop

```python
from typing import List

def printNos(x: int, start: int = 1) -> List[int]: 
    # Write your code here
    if start > x:
        return []
    return [start] + printNos(x, start + 1)
```

## N to 1 without using Loop

```python
from typing import List

def printNos(x: int, acc : List[int] = []) -> List[int]:
    if x == 0:
        return acc
    return printNos(x - 1, acc + [x])
```

```python
printNos(10)
```

## N to 1 without loop

```python
from typing import List

def printNos(x: int, acc : List[int] = []) -> List[int]:
    if x == 0:
        return acc
    return printNos(x - 1, acc + [x])
```

```python
printNos(11)
```

```python
def sumFirstN(n: int, sum : int = 0) -> int:
    if n == 0:
        return sum
    sum += n
    return sumFirstN(n - 1, sum)
```

```python
sumFirstN(11)
```

```python
# Will change this approach to use recursion

def factorialNumbers(n: int) -> List[int]:
    factorials = [1]
    i = 1
    while True:
        i += 1
        next_factorial = factorials[-1] * i
        if next_factorial > n:
            break
        factorials.append(next_factorial)
    return factorials
```

```python
factorialNumbers(7)
```
