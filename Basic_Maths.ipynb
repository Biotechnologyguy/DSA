## Number of digits that evenly divides n

```python
def countDigits(n: int) -> int:
    orgnl_num  = n
    digi_count = 0
    while(n > 0):
        last_digit = n % 10
        if last_digit != 0 and orgnl_num % last_digit == 0:
            digi_count += 1
        n = int(n / 10)
    return digi_count
```

```python
countDigits(336)
```

## Reverse a number

```python
def reverse(x):
        """
        :type x: int
        :rtype: int
        """
        INT_MAX = 2**31 - 1
        
        sign = -1 if x < 0 else 1
        x_abs = abs(x)
        
        # Reverse the digits of the number
        reversed_x = 0
        while x_abs != 0:
            pop = x_abs % 10
            x_abs = int(x_abs / 10)
            
            # Check for overflow
            if reversed_x > INT_MAX // 10 or (reversed_x == INT_MAX // 10 and pop > INT_MAX % 10):
                return 0
            reversed_x = reversed_x * 10 + pop
        
        # Apply the sign to the reversed number
        return sign * reversed_x
```

```python
reverse(x=32000)
```

## Check Palindrome

```python
def isPalindrome(x):
    # string = str(x)
    # if string == string[::-1]:
    #     return True
    # else:
    #     return False
    num = x
    reversed = 0
    while(x > 0):
        reversed = reversed * 10 + x % 10
        x //= 10
    return num == reversed
```

```python
isPalindrome(-121)
```

```python
isPalindrome(121)
```

```python
isPalindrome(3113)
```

## Check armstrom

```python
num = int(input())
power = len(str(num))
dup = num
num_sum = 0
while(num > 0):
    num_sum += (num % 10)**power
    num //= 10

print("true" if num_sum == dup else "false")
```

## fibonacci number

```python
def fib(n):
    """
    :type n: int
    :rtype: int
    """
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a
```

```python
fib(11)
```

## Calculate GCD

##### GCD(a, b) = GCD(a - b, b) where a > b


The Euclidean algorithm is based on the principle that the greatest common divisor (GCD) of two numbers \(a\) and \(b\), where \(a > b\), is the same as the GCD of their difference \(a - b\) and the smaller number \(b\). In other words, \(GCD(a, b) = GCD(a - b, b)\). The process involves iteratively reducing the larger number by subtracting the smaller number from it until the larger number becomes less than the smaller number. At this point, the roles of the larger and smaller numbers are swapped, and the process repeats until one of the numbers becomes zero. At that stage, the non-zero number is the GCD of the original pair of numbers.

While this method is correct, it can be quite slow if the numbers are large because it may require many subtraction steps to reduce \(a\) to a number smaller than \(b\). A more time-efficient approach is to use the modulo operator (%) which effectively performs multiple subtractions in one step. By replacing the difference \(a - b\) with the remainder of the division \(a \% b\), the algorithm significantly reduces the number of steps required to find the GCD. Thus, \(GCD(a, b) = GCD(b, a \% b)\) when \(a \% b\) is not zero. This modification ensures that the algorithm runs in O(log(min(a, b))) time, making it much faster, especially for large numbers.

```python
def calcGCD(n:int, m: int) -> int:
    # while m:
    #     n, m = m, n % m
    # return n
    while(n > 0 and m > 0):
        if n > m:
            n %= m
        else:
            m %= n
        if n == 0:
            return m
        if m == 0:
            return n
```

```python
calcGCD(52, 10)
```
