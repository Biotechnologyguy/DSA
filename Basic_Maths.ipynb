## Number of digits that evenly divides n

```python
def countDigits(n: int) -> int:
    orgnl_num  = n
    digi_count = 0
    while(n > 0):
        last_digit = n % 10
        if last_digit != 0 and orgnl_num % last_digit == 0:
            digi_count += 1
        n = int(n / 10)
    return digi_count
```

```python
countDigits(336)
```

## Reverse a number

```python
def reverse(x):
        """
        :type x: int
        :rtype: int
        """
        INT_MAX = 2**31 - 1
        
        sign = -1 if x < 0 else 1
        x_abs = abs(x)
        
        # Reverse the digits of the number
        reversed_x = 0
        while x_abs != 0:
            pop = x_abs % 10
            x_abs = int(x_abs / 10)
            
            # Check for overflow
            if reversed_x > INT_MAX // 10 or (reversed_x == INT_MAX // 10 and pop > INT_MAX % 10):
                return 0
            reversed_x = reversed_x * 10 + pop
        
        # Apply the sign to the reversed number
        return sign * reversed_x
```

## Check Palindrome

```python
def isPalindrome(x):
    # string = str(x)
    # if string == string[::-1]:
    #     return True
    # else:
    #     return False
    num = x
    reversed = 0
    while(x > 0):
        reversed = reversed * 10 + x % 10
        x //= 10
    return num == reversed
```

```python
isPalindrome(-121)
```

```python
isPalindrome(121)
```

```python
isPalindrome(3113)
```

## Check armstrom

```python
num = int(input())
power = len(str(num))
dup = num
num_sum = 0
while(num > 0):
    num_sum += (num % 10)**power
    num //= 10

print("true" if num_sum == dup else "false")
```

## fibonacci number

```python
def fib(self, n):
    """
    :type n: int
    :rtype: int
    """
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a
```
